PROJECT BACKGROUND


The project question:  
    
Is there a relationship between fuel poverty (FP) and the relative Energy Performance Certificate (EPC) scores of English Local Authority Districts (LAD)  ?

The potential ‘public good’ insight that might be obtained is that if such a relationship is established and it is believed that there is a causal relationship
(e.g. lower average EPC LADs contribute to higher FP )  then the national Government , regional authority or LAD might target resources to improve the EPC ratings of those relatively lower EPC LADs and thereby be more efficient in reducing FP.


Readership

The initial hypothesis is that lower energy efficiency in a LAD is one of the contributing factors that leads to a higher level of 
fuel poverty(FP). If such a causal link is credible then policymakers at a national or local level might reduce FP by targetting 
support to improve the energy efficiency of lowly rated LADs ('the public good').


DEFINITIONS


FUEL POVERTY (FP) 

The UK’s Department for Energy Security & Net Zero (DESNZ) determines which households are classified as 
FP by reference to the Low Income Low Energy Efficiency Rating , that is if the household has: 

"a property with an energy efficiency rating of band D, E, F or G" 
and …its disposable income (income after housing costs (AHC) and energy costs) 
would be below the poverty line" [FN1]  

"The poverty line (income poverty) is defined as an equivalised disposable income of less 
than 60% of the national median" [FN 2]  



KEY DRIVERS FOR FP

DESNZ[FN 3] sets out the key drivers of FP  as: 
    
- Energy Efficiency

- Energy Prices 

- Incomes 


The project question focuses on the energy efficiency input rather than energy prices and incomes 
but recognises that whilst energy prices and incomes (via minimum wage or benefit policies) are normally 
the concern of  central rather than local government there are large regional and intra- regional LAD 
differences  in incomes so FP in LAD is a combination of factors . Separating the impact of 
uneven income distribution within that mix is beyond this paper. 



ENERGY PERFORMANCE CERTIFICATES (EPC) AND POTENTIAL EPC (PEPC) SCORES 

EPCs contain scores and ratings for domestic residences in the UK ,  the Department for Communities
and Local Government (DCLG) explains their purpose as follows :

"The EPC will provide an energy efficiency rating from A to G, where A is very efficient 
and G is the least efficient. The better the rating, the more energy efficient the building is, 
and the lower the fuel bills are likely to be….

Each energy efficiency rating is based on the characteristics of the building itself 
(the fabric) and its services (such as heating, ventilation and lighting) …

The EPC includes recommendations to help owners and occupiers to improve the energy efficiency of a building. ….
The potential rating shown on the EPC is based on all cost-effective recommendations being implemented" [FN 4]


Alongside the A- G ratings there is a related score out of 100 so the difference in 
scores between two residential buildings can be calculated on a very granularly 
basis i.e. : 
    
"Energy Performance Certificate (EPC) provides a numerical scale of the theoretical cost of energy use in a 
residential property. This energy performance rating is standardised to a scale of zero (high cost) to 
100 (minimal cost), which is frequently used as a measure of energy use of residential buildings in the UK" [FN 5]

Alongside the EPC score there is a PEPC score that looks at the potential score for the residential property
if all the ‘cost-effective’ measures were taken. It does not take into account all possible measures  since 
that might make every potential score 100 ( but at a very high cost).

1.The information on moving into and out from LADs related to changes in EPC and PEPC scores and 
floor area (in square metres )  contains both mean changes and median changes for Great Britain 
( i.e. England , Scotland and Wales) . The project aims to explore whether there is a relationship
between LAD’s relative EPC position and its FP. The FP data is for England only and therefore in both
datasets ( moving in and moving out) the non English ( i.e. Scotland and Wales LADs are removed) and the
median information is deemed to be not  particularly informative (given that the mean information is retained)
for the additional columns it would require. The code therefore aims to leave only English LADs 
(309 rather than 363) and remove the three median related  columns.

Arguably the ‘mean_PEPC change’ and  ‘mean_area_change’ are candidates for deletion, but 
whilst the initial proposition is related to EPC score differences , continuing the analysis 
including PEPCs and areas may add insights. Also because the two datasets have identical column 
headings future confusion can be avoided by adding a tag to the columns reflecting whether the data
is for moving into an LAD (‘mi’) or alternatively moving out (‘mo’). Whilst making that change 
the columns names for the variables were also shortened to aid readability of the notebook
(as more columns are added in merging datasets later).   The first merging is for the mi and mo
datasets using the common column containing the area code ( ‘laua21’).


2.The ONS geographical information is dealt with in a similar way the csv file contains many columns 
relating to the geographical elements of the LADs but the required columns are only the area codes
( labelled “LAD21CD’ in this dataset) and Area name (‘LAD21NM) . The geometry shp file contains only 
one column i.e geometry. The two datasets come from the same ONS suite , so there is an expectation
that the rows in each are reflecting the same LADs ( Area code and name in index 0 matches geometry in index 0).
Once merged this assumption is tested by checking the name of identifiable LADs to their now added geometry. 
In the 3 tested cases this is proven to be correct .


3.The two already merged files are then merged again to add the EPC etc information to the over 
lapping geometry and LAD name information ( i.e. English LADs only. It is now possible to review  not only
the statistics and correlation for the six variables but also reflect these for each LAD in England spatially. 
Unsurprisingly the correlation coefficents between each corresponding mean for mi and mo are highly negative 
( all less than -0.9) and that allows for a further simplification of combining the mi and mo into one 
amounts for the three variables ( to maintain score comparability and result in high scores reflecting the 
LADs likely to have the more energy efficient housing stock , the equation for combining is (mi-mo)/2).
